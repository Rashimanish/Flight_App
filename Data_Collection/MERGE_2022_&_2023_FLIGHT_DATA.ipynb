```python
import pandas as pd
import os

path = '/content/Data'
output_path_2022 = '/content/Data/flight_data_2022.csv'
output_path_2023 = '/content/Data/flight_data_2023.csv'
output_path_combine = '/content/Data/combined_flightdata.csv'

# Load the 2022 and 2023 data
df_2022 = pd.read_csv(output_path_2022)
df_2023 = pd.read_csv(output_path_2023)

# Concatenate the DataFrames for 2022 and 2023 data
combined_df = pd.concat([df_2022, df_2023], ignore_index=True)

# Save the concatenated DataFrame to a CSV file
combined_df.to_csv(output_path_combine, index=False)

# Display the first few rows of the combined DataFrame
print(combined_df.head())

print("\nNumber of rows and columns in the combined DataFrame:")
print(combined_df.shape)


print("\nCombined DataFrame info:")
print(combined_df.info())

print("\nCombined DataFrame descriptive statistics:")
print(combined_df.describe())
```
